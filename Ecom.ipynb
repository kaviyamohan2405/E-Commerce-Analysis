```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
# Load the JSON dataset into a DataFrame
df = pd.read_json('D:/Job Fair Project/E-Commerce Analysis/fashion_products_dataset.json')
```

```python
df1=pd.DataFrame(df)
```

```python
df1
```

```python
df1['product_details']
```

```python
import pandas as pd

# Function to extract data from product_details
def extract_data(details):
    extracted_data = {}
    for detail in details:
        for key, value in detail.items():
            extracted_data[key] = value
    return extracted_data

# Assuming df1 is your DataFrame with 30,000 rows
# Extract data from product_details column
extracted_data = df1['product_details'].apply(extract_data)

# Convert the extracted data into a DataFrame
df_extracted = pd.DataFrame(extracted_data.tolist())

# Display the extracted data
print(df_extracted)
```

```python
# Concatenate df1 with df_extracted
df_combined = pd.concat([df1, df_extracted], axis=1)

# Display the combined DataFrame
print(df_combined)
```

```python
overall=pd.DataFrame(df_combined)
```

```python
overall
```

```python
overall.columns
```

```python
# Assuming your DataFrame is named df
df_filled = overall.fillna(value=0)  # Fill NaN values with 0
```

```python
import pandas as pd
import numpy as np

# Assuming your DataFrame is named df
df_filled.replace('', np.nan, inplace=True)  # Replace empty strings with NaN
df_cleaned = df_filled.dropna()  # Remove rows containing NaN values
```

# Identifying popular product categories and brands:

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='category', data=df_cleaned)
plt.title('Number of Products by Category')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='brand', data=df_cleaned, order=df_cleaned['brand'].value_counts().iloc[:10].index)
plt.title('Top 10 Brands by Product Count')
plt.xlabel('Brand')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

# Analyzing pricing trends and discount strategies

```python
plt.figure(figsize=(10, 6))
sns.histplot(df_cleaned['actual_price'], kde=True)
plt.title('Distribution of Actual Prices')
plt.xlabel('Actual Price')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df_cleaned['discount'], bins=30)
plt.title('Discount Frequency Distribution')
plt.xlabel('Discount (%)')
plt.ylabel('Number of Products')
plt.show()
```

# Understanding seller behavior and performance

```python
plt.figure(figsize=(12, 8))
sns.countplot(y='seller', data=df_cleaned, order=df_cleaned['seller'].value_counts().iloc[:10].index)
plt.title('Top 10 Sellers by Product Count')
plt.xlabel('Number of Products')
plt.ylabel('Seller')
plt.show()
```

# Investigating customer preferences and product satisfaction

```python
plt.figure(figsize=(10, 6))
sns.histplot(df_cleaned['average_rating'], bins=20, kde=True)
plt.title('Distribution of Average Ratings')
plt.xlabel('Average Rating')
plt.ylabel('Count')
plt.show()
```

# Exploring correlations between various attributes

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='actual_price', y='average_rating', hue='category', data=df_cleaned)
plt.title('Actual Price vs. Average Rating by Category')
plt.xlabel('Actual Price')
plt.ylabel('Average Rating')
plt.legend(title='Category', loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```
